```python
def ler_cotacoes_investing( caminho_arquivo_ativo_initeresse_input, caixa_inicial ):
  caminho_arquivo = caminho_arquivo_ativo_initeresse_input + '.csv'
  dataframe_cotacao = pd.read_csv(caminho_arquivo)  
  dataframe_cotacao = dataframe_cotacao.sort_index(ascending=False)
  dataframe_cotacao = dataframe_cotacao.reset_index()
  dataframe_cotacao = dataframe_cotacao.filter(items=['Data', 'Último'])
  dataframe_cotacao = dataframe_cotacao.set_index('Data')
  dataframe_cotacao = dataframe_cotacao.apply(lambda x: x.str.replace(',','.'))
  dataframe_cotacao = pd.DataFrame(dataframe_cotacao)
  dataframe_cotacao['Último'] = dataframe_cotacao['Último'].astype(float)
  dataframe_cotacao.columns = ['Fechamento_ajustado_diario']

  dataframe_cotacao['Retorno_diario'] = dataframe_cotacao['Fechamento_ajustado_diario'].pct_change(1)
  dataframe_cotacao['Retorno_diario_normalizado'] = dataframe_cotacao['Fechamento_ajustado_diario']/dataframe_cotacao.iloc[0]['Fechamento_ajustado_diario']
  dataframe_cotacao['Valor_diario_da_carteira'] = dataframe_cotacao['Retorno_diario_normalizado']*caixa_inicial

  #return dataframe_cotacao[-1000:]
  return dataframe_cotacao
```

**procurar fonte na literatura sobre como calcular o beta, e o alpha
o beta foi calculo como sendo a correlação entre o retorno diario do portifolio de interesse, contra o retorno diario do portifolio de benchmark.

Já o alpha foi calculado como sendo a taxa de retorno do porttaxa_de_retorno_portifolio - taxa_livre_de_risco - beta*(taxa_de_retorno_benchmark - taxa_livre_de_risco)

a taxa livre de risco foi definida arbitrariamente como sendo 48% no periodo, ou seja, 4% no periodo total de 10 anos

```python
from scipy import stats


def calculo_alpha_beta(dataframe_ativo_de_interesse, dataframe_ativo_de_benchmark ):


  (beta, alpha) = stats.linregress(dataframe_ativo_de_benchmark['Retorno_diario'].values[1:],
              dataframe_ativo_de_interesse['Retorno_diario'].values[1:])[0:2]
  
  taxa_de_retorno_benchmark = (dataframe_ativo_de_benchmark['Valor_diario_da_carteira'][-1]/dataframe_ativo_de_benchmark['Valor_diario_da_carteira'][0])
  taxa_de_retorno_portifolio = (dataframe_ativo_de_interesse['Valor_diario_da_carteira'][-1]/dataframe_ativo_de_interesse['Valor_diario_da_carteira'][0])

  taxa_livre_de_risco = 1.2273

  alpha = taxa_de_retorno_portifolio - taxa_livre_de_risco - beta*(taxa_de_retorno_benchmark - taxa_livre_de_risco) 


  return (round(beta, 4),round(alpha, 5))
```

** procurar uma fonte na literatura para esse calculo de indice sharp
atualmente o calculo esta usando o crescimento da carteira, dividido pelo desvio padrão do retorno diario.

Então para caso de o desvio padrão dos rendimentos diarios seja alta, o indice sharp sera menor. 
Por fim, quanto maior rendimento, com menor volatilidade, mais positivo será o indice sharp.

```python
def calculo_sharp(dataframe_ativo_interesse):

  #Sharpe_Ratio = dataframe_ativo_interesse['Retorno_diario'].mean()/dataframe_ativo_interesse['Retorno_diario'].std()
  
  retorno_percentual_do_portifolio = -1+(dataframe_ativo_interesse['Valor_diario_da_carteira'][-1]/dataframe_ativo_interesse['Valor_diario_da_carteira'][0])
  Sharpe_Ratio_Total = (retorno_percentual_do_portifolio)/dataframe_ativo_interesse['Retorno_diario'].std()

  Anualizado_Sharp_Ratio = 0
  

  return Sharpe_Ratio_Total, Anualizado_Sharp_Ratio
```

```python
def calculo_percentual_final(dataframe_ativo_interesse, dataframe_ativo_shannon_demon):
  
  lucro_percentual_final_buy_and_hold = -100+(dataframe_ativo_interesse['Valor_diario_da_carteira'][-1]/dataframe_ativo_interesse['Valor_diario_da_carteira'][0])*100
  lucro_percentual_final_shannon = -100+(dataframe_ativo_shannon_demon['Valor_diario_da_carteira'][-1]/dataframe_ativo_shannon_demon['Valor_diario_da_carteira'][0])*100


  return lucro_percentual_final_buy_and_hold, lucro_percentual_final_shannon
```

```python
def calculo_volatilidade(dataframe_ativo_interesse, dataframe_ativo_shannon_demon):
  
  volatiliidade_buy_and_hold = dataframe_ativo_interesse['Retorno_diario'].std()
  volatiliidade_shannon = dataframe_ativo_shannon_demon['Retorno_diario'].std()


  return volatiliidade_buy_and_hold, volatiliidade_shannon
```

```python
def execucao_de_shannons_rebalance( dataframe_ativo_de_interesse, treshold_rebalanco, caixa_inicial ):

  balanco = 1
  caixa_livre = caixa_inicial/2
  caixa_em_acoes = caixa_livre
  valor_max_local_do_caixa = 0
  max_drawn_down = 0
  max_drawn_down_percentual = 0


  quantidade_de_rebalanco = 0
  numero_do_dia_do_ultimo_rebalaco = 0

  lista_shannons_rebalanco = []

  lista_aux_lista_shannons_rebalanco = [dataframe_ativo_de_interesse.index[0], caixa_inicial]
  lista_shannons_rebalanco.append(lista_aux_lista_shannons_rebalanco)


  for numero_do_dia in range(0, -1+len(dataframe_ativo_de_interesse['Fechamento_ajustado_diario'])):




    stock_price_now = dataframe_ativo_de_interesse['Fechamento_ajustado_diario'][numero_do_dia+1]
    stock_price_before = dataframe_ativo_de_interesse['Fechamento_ajustado_diario'][numero_do_dia]
    caixa_em_acoes = caixa_em_acoes + caixa_em_acoes*(stock_price_now-stock_price_before)/stock_price_before
    balanco = caixa_em_acoes/caixa_livre

    lista_aux_lista_shannons_rebalanco = [dataframe_ativo_de_interesse.index[numero_do_dia+1], caixa_em_acoes+caixa_livre]
    lista_shannons_rebalanco.append(lista_aux_lista_shannons_rebalanco)



    if (balanco < 1 + treshold_rebalanco) and (1 - treshold_rebalanco < balanco) :
        pass
    else:

        '''
        intervalo_de_dias = numero_do_dia - numero_do_dia_do_ultimo_rebalaco
        numero_do_dia_do_ultimo_rebalaco = numero_do_dia
        intervalo_de_dias_por_opcao_mensal = intervalo_de_dias/22
        premio_por_opcao = 2*(0.020+0.008)*intervalo_de_dias_por_opcao_mensal*caixa_em_acoes*0.1
        '''


        quantidade_de_rebalanco = quantidade_de_rebalanco + 1
        balanco = caixa_em_acoes/caixa_livre
        #print('preço da acao hoje :', stock_price_now, 'xx preço da ação ontem :', stock_price_before, 'xx caixa em ações :', caixa_em_acoes,'xx caixa livre :', caixa_livre, 'xx balanço :', balanco )
        caixa_para_rebalanco = caixa_livre - caixa_em_acoes
        #caixa_livre = premio_por_opcao + caixa_livre - caixa_para_rebalanco/2 
        taxa_bolsa = 0.0050*caixa_para_rebalanco #https://www.b3.com.br/pt_br/produtos-e-servicos/tarifas/listados-a-vista-e-derivativos/renda-variavel/tarifas-de-acoes-e-fundos-de-investimento/a-vista/
        caixa_livre = caixa_livre - taxa_bolsa - caixa_para_rebalanco/2 
        caixa_em_acoes = caixa_em_acoes + caixa_para_rebalanco/2
        balanco = caixa_em_acoes/caixa_livre
        #print('preço da acao hoje :', stock_price_now, 'xx preço da ação ontem :', stock_price_before, 'xx caixa em ações :', caixa_em_acoes,'xx caixa livre :', caixa_livre, 'xx balanço :', balanco )
        
        
    

    if valor_max_local_do_caixa < caixa_livre + caixa_em_acoes:
        valor_max_local_do_caixa = caixa_livre + caixa_em_acoes
    elif caixa_livre + caixa_em_acoes < valor_max_local_do_caixa:
        if max_drawn_down < valor_max_local_do_caixa - (caixa_livre + caixa_em_acoes):
            
            max_drawn_down = valor_max_local_do_caixa - (caixa_livre + caixa_em_acoes)
            max_drawn_down_percentual = (valor_max_local_do_caixa - (caixa_livre + caixa_em_acoes))/valor_max_local_do_caixa

  

  dataFrame_shannons_rebalanco = pd.DataFrame(lista_shannons_rebalanco)
  dataFrame_shannons_rebalanco.columns = ['Data','Valor_diario_da_carteira']
  dataFrame_shannons_rebalanco = dataFrame_shannons_rebalanco.set_index('Data')


  retorno_acumulado = 100 * ( dataFrame_shannons_rebalanco [ 'Valor_diario_da_carteira' ] [-1 ] / dataFrame_shannons_rebalanco [ 'Valor_diario_da_carteira' ] [ 0 ] -1)
  dataFrame_shannons_rebalanco['Retorno_diario'] = dataFrame_shannons_rebalanco['Valor_diario_da_carteira'].pct_change(1)
  '''
  print('xxxxxxxxxxxxxxx')
  print(dataFrame_shannons_rebalanco['Valor_diario_da_carteira'][-1])
  print(dataframe_ativo_de_interesse['Valor_diario_da_carteira'][-1])
  '''

  return dataFrame_shannons_rebalanco, quantidade_de_rebalanco
    
```
